```python
from numpy.core.fromnumeric import size
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv("benchmark.csv")
```

```python
fig = px.scatter(df, x="fps", y="top1",
                 color="model", # Species 열의 값에 따라서 색깔 표현
                 hover_data=['param_count'], # 참고할 데이터 추가
                 title='PyTorch Backbone Benchmark - Scatter Plot(Y: top1 acc, X: fps)' # 그래프 타이틀 지정
                )
fig.show()
```

```python
fig = px.scatter(df, x="param_count", y="top1",
                 color="model", # Species 열의 값에 따라서 색깔 표현
                 hover_data=['fps'], # 참고할 데이터 추가
                 title='PyTorch Backbone Benchmark - Scatter Plot(Y: top1 acc, X: #param)' # 그래프 타이틀 지정
                )
fig.show()
```
